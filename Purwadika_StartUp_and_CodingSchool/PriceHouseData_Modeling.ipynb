```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["figure.figsize"]= 20,10
%matplotlib inline
```

```python
#membaca dataset hasil akhir data processing
data11 = pd.read_csv('PriceHouseData_OutlierRemoval.csv')
data11.head()
```

```python
#pada tahapan ini kita akan membuat model machine learning kemudian menggunakan k-fold cross validation dan grid cv agar hadir dengan alogoritm dan parameter yang terbaik
```

```python
#namun sebelumnya, model machine learning hanya dapat membaca informasi numeric sedangkan location feature berisi string informasi
#sehingga kita butuh teknik feature engineering yakni one-hot encoding untuk mentranformasi string to numeric information
```

```python
#one-hot coding jugs sering dikenal sebagai dummies data
dummies = pd.get_dummies(data11.location)
dummies.head()
```

```python
data12 = pd.concat([data11,dummies.drop('other', axis='columns')], axis='columns')
data12.head()
```

```python
data13 = data12.drop('location', axis='columns')
data13.head()
```

```python
data13.shape
```

```python
#kemudian kita akan menentukan dependent variable dan independent variable. pada dataset price adalah dependent variable
```

```python
#variable X mengandung semua independet variable
X = data13.drop('price', axis='columns')
X.head()
```

```python
#variable Y mengadung dependent variable
y= data13.price
y.head()
```

```python
#kemudian kita akan membagi dataset menjadi data training dan data test dimana 80% dataset sebagai data training dan sisinya untuk data test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression() #model regression
lr_clf.fit(X_train, y_train) # fitting model
lr_clf.score(X_test,y_test) #evaluate score model
#diperoleh score evaluasi model sekitar 80 persen. hasil ini menunjukan model yang bagus
```

```python
#dalam proses data science ada baiknya kita menggunakan beberapa model, parameter agar mendapatkan model yang optimal
```

```python
#sebelum kita membuat beberapa model kita akan melakukan k-fold cross validation
from sklearn.model_selection import ShuffleSplit  
from sklearn.model_selection import cross_val_score
# merendom sample kita sehingga setiap fold mempunyai distribusi yang sama
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(),X,y,cv=cv)
#kita peroleh score evaluasi dikisaran 80% mengunakan model linear Regression
```

```python
#kemudian kita akan mencoba beberapa model regression seperti lasso dan decision regresi.
#kita butuh mencoba beberapa algoritma dan parameter agar mendapatkan score yang terbaik
#oleh karena itu kita akan mengunakan metode grid search CV yang merupakn good API yang disediakan sklearn
#yang mana dapat menjalankan berbagai algoritma and parameter
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
#dapat kita lihat bahwa linear regression menghasilkan model terbaik
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
